```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_param = 2    # Loss aversion parameter
gamma = 0.9        # Curvature parameter
c = 0              # Reference point (benchmark level)

# Define the S-shaped utility function
def utility_function(x, c=0, lambda_param=2, gamma=0.9):
    """
    S-shaped utility function from Tversky and Kahneman (1992)

    u(x) = (x-c)^gamma      if x >= c
    u(x) = -lambda(c-x)^gamma  if x < c
    """
    result = np.zeros_like(x)

    # Gains domain (x >= c)
    gains_mask = x >= c
    result[gains_mask] = (x[gains_mask] - c) ** gamma

    # Losses domain (x < c)
    losses_mask = x < c
    result[losses_mask] = -lambda_param * (c - x[losses_mask]) ** gamma

    return result

# Define the risk aversion coefficient
def risk_aversion_coefficient(x, c=0, gamma=0.9):
    """
    Arrow-Pratt coefficient of absolute risk aversion

    alpha(x) = (1-gamma)/(x-c)     if x > c
    alpha(x) = (gamma-1)/(c-x)     if x < c
    """
    result = np.zeros_like(x)

    # Gains domain (x > c)
    gains_mask = x > c
    result[gains_mask] = (1 - gamma) / (x[gains_mask] - c)

    # Losses domain (x < c)
    losses_mask = x < c
    result[losses_mask] = (gamma - 1) / (c - x[losses_mask])

    return result

# Generate data points
x_range = np.linspace(-5, 5, 1000)

# Exclude points very close to the reference point to avoid division by zero
epsilon = 0.01
x_gains = x_range[x_range > epsilon]
x_losses = x_range[x_range < -epsilon]
x_plot = np.concatenate([x_losses, x_gains])

# Calculate utility values
u_values = utility_function(x_plot, c, lambda_param, gamma)

# Calculate risk aversion values
alpha_values = risk_aversion_coefficient(x_plot, c, gamma)

# Create the plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot 1: Utility Function
ax1.plot(x_losses, utility_function(x_losses, c, lambda_param, gamma),
         'r-', linewidth=3, label='Losses (x < 0)', color='#ff6b6b')
ax1.plot(x_gains, utility_function(x_gains, c, lambda_param, gamma),
         'b-', linewidth=3, label='Gains (x > 0)', color='#4ecdc4')

# Add reference lines
ax1.axhline(y=0, color='gray', linestyle='--', alpha=0.7, linewidth=1)
ax1.axvline(x=0, color='black', linestyle='--', alpha=0.8, linewidth=2, label='Reference Point')

# Formatting for utility plot
ax1.set_xlabel('Wealth Change (x)', fontsize=12)
ax1.set_ylabel('Utility u(x)', fontsize=12)
ax1.set_title(f'S-Shaped Utility Function (={lambda_param}, ={gamma})', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend(fontsize=11)
ax1.set_xlim(-5, 5)
ax1.set_ylim(-4, 3)

# Add annotations
ax1.annotate('Risk Averse\n(Concave)', xy=(2, 1.5), xytext=(3, 2.5),
            arrowprops=dict(arrowstyle='->', color='#4ecdc4', lw=2),
            fontsize=11, ha='center', color='#4ecdc4', fontweight='bold')
ax1.annotate('Risk Seeking\n(Convex)', xy=(-2, -1.5), xytext=(-3, -2.5),
            arrowprops=dict(arrowstyle='->', color='#ff6b6b', lw=2),
            fontsize=11, ha='center', color='#ff6b6b', fontweight='bold')

# Plot 2: Risk Aversion Coefficient
ax2.plot(x_losses, risk_aversion_coefficient(x_losses, c, gamma),
         'r-', linewidth=3, label='Losses (Risk Seeking)', color='#ff6b6b')
ax2.plot(x_gains, risk_aversion_coefficient(x_gains, c, gamma),
         'b-', linewidth=3, label='Gains (Risk Averse)', color='#4ecdc4')

# Add reference lines
ax2.axhline(y=0, color='gray', linestyle='--', alpha=0.7, linewidth=1)
ax2.axvline(x=0, color='black', linestyle='--', alpha=0.8, linewidth=2, label='Reference Point')

# Formatting for risk aversion plot
ax2.set_xlabel('Wealth Change (x)', fontsize=12)
ax2.set_ylabel('Risk Aversion (x)', fontsize=12)
ax2.set_title(f'Arrow-Pratt Coefficient (x) = (1-)/|x-c|  sign(x-c)', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.legend(fontsize=11)
ax2.set_xlim(-5, 5)
ax2.set_ylim(-1, 1)

# Add annotations for risk aversion plot
ax2.annotate(' > 0\n(Risk Averse)', xy=(2, 0.05), xytext=(3.5, 0.3),
            arrowprops=dict(arrowstyle='->', color='#4ecdc4', lw=2),
            fontsize=11, ha='center', color='#4ecdc4', fontweight='bold')
ax2.annotate(' < 0\n(Risk Seeking)', xy=(-2, -0.05), xytext=(-3.5, -0.3),
            arrowprops=dict(arrowstyle='->', color='#ff6b6b', lw=2),
            fontsize=11, ha='center', color='#ff6b6b', fontweight='bold')

# Adjust layout and display
plt.tight_layout()
plt.show()

# Print some key values for verification
print("=" * 60)
print("S-SHAPED UTILITY FUNCTION ANALYSIS")
print("=" * 60)
print(f"Parameters:  = {lambda_param},  = {gamma}, c = {c}")
print()

# Calculate some specific values
test_points = [-3, -1, 1, 3]
print("Sample calculations:")
print("x\t\tu(x)\t\t(x)")
print("-" * 40)
for x in test_points:
    u_val = utility_function(np.array([x]), c, lambda_param, gamma)[0]
    alpha_val = risk_aversion_coefficient(np.array([x]), c, gamma)[0]
    print(f"{x:6.1f}\t\t{u_val:8.4f}\t\t{alpha_val:8.4f}")

print()
print("Key Properties:")
print(" Risk averse for gains (x > 0): (x) > 0")
print(" Risk seeking for losses (x < 0): (x) < 0")
print(" Decreasing absolute risk aversion: |(x)| decreases as |x| increases")
print(" Loss aversion: utility function steeper for losses than gains")
print(" Reference point at x = 0 creates a 'kink' in the utility function")

# Optional: Save the plots
plt.savefig('s_shaped_utility_analysis.png', dpi=300, bbox_inches='tight')
print("\nPlot saved as 's_shaped_utility_analysis.png'")
```
